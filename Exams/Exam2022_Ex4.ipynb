a) Explain the concept and implementation of Maximum Common Subgraph (MCS) and apply it to G1 and G2.

The Maximum Common Subgraph (MCS) is a graph theory concept and algorithm used to find the largest common subgraph between two given graphs, G1 and G2. In other words, it identifies the largest subgraph that can be found in both G1 and G2. This concept is widely used in various fields, including chemistry, biology, computer science, and network analysis, where it can help in comparing and analyzing the structural similarities between different graphs.

Implementation of the MCS algorithm typically involves the following steps:

Input: You start with two input graphs, G1 and G2, for which you want to find the maximum common subgraph.

Enumeration of Subgraphs: Enumerate all possible subgraphs of G1 and G2. This can be an exhaustive process, especially for larger graphs, but various algorithms and strategies can be employed to make this step more efficient.

Comparison: For each pair of subgraphs (one from G1 and one from G2) generated in step 2, compare them to check if they are isomorphic. Isomorphic subgraphs have the same structure; that is, they have the same number of nodes and edges and the same connectivity.

Largest Common Subgraph: Keep track of the largest isomorphic subgraph found during the comparison process. This subgraph represents the maximum common subgraph between G1 and G2 found so far.

Repeat: Continue the enumeration and comparison process until all possible subgraphs have been considered or until you have a termination condition (e.g., when you are sure you have found the largest common subgraph).

Output: The largest common subgraph found is the result of the MCS algorithm.

b) Calculate value of max-normalized distance $Mdist(G1, G2) = 1 - \dfrac{|MCS(G1, G2)|} {max{|G1|,|G2|}}$ and discuss its properties.


Union-normalized distance: The distance measure lies in the range of (0, 1), and is
also shown to be a metric. The union-normalized measure UDist(G1, G2) is defined
as follows:

$$
UDIST(G_1,G_2) = 1 - \frac{\mid MCS(G_1,G_2) \mid}{ \mid G_1 \mid +\mid G_2 \mid - \mid MCS(G_1,G_2) \mid}
$$

This measure is referred to as the union-normalized distance because the denominator
contains the number of nodes in the union of the two graphs. A different way of
understanding this measure is that it normalizes the number of non-matching nodes
U(G1, G2) between the two graphs (unnormalized measure) with the number of nodes
in the union of the two graphs.

$$
UDist(G_1, G_2) = \frac{\text{Non-matching nodes between $G_1$ and $G_2$}}{\text{Union size of $G_1$ and $G_2$}}
$$

One advantage of this measure is that it is intuitively easier to interpret. Two perfectly
matching graphs will have a distance of 0 from one another, and two perfectly nonmatching graphs will have a distance of 1.


Max-normalized distance: This distance measure also lies in the range (0, 1). The max-normalized distance $MDist(G_1, G_2)$ between two graphs $G_1$ and $G_2$ is defined as follows:

$$ MDist(G_1, G_2) = 1 − \frac{|MCS(G_1, G_2)|}{max\{|G1|, |G2| \}}$$

The main difference from the union-normalized distance is that the denominator is normalized by the maximum size of the two graphs. This distance measure is a metric because it satisfies the triangle inequality. The measure is also relatively easy to interpret. Two perfectly matching graphs will have a distance of 0 from one another, and two perfectly non-matching graphs will have a distance of 1.

e) How could you use frequent subgraphs for implementing a distance measure between two graphs?


Frequent subgraphs can be used to implement a distance measure between two graphs by quantifying the structural similarity or dissimilarity between them based on the presence and frequency of common subgraphs. This approach is particularly useful in applications such as graph database querying, pattern recognition, and similarity assessment in graph data. Here's how you can use frequent subgraphs to implement a distance measure between two graphs:

Frequent Subgraph Mining:

First, you need to perform frequent subgraph mining on both of the input graphs (Graph A and Graph B). This involves finding all frequent subgraphs in each graph based on a certain support threshold. Frequent subgraphs are subgraphs that occur frequently enough in the data according to a predefined minimum support value. Various algorithms like Apriori-based algorithms, gSpan, or FSG can be used for this purpose.
Identify Common Frequent Subgraphs:

After mining frequent subgraphs in both graphs, identify the subgraphs that are common to both graphs. These common subgraphs represent structural patterns shared between Graph A and Graph B.
Calculate Distance Measure:

The distance measure between the two graphs can be defined based on the presence and frequency of common subgraphs. Several distance metrics can be used, including:

Jaccard Distance: Calculate the Jaccard similarity coefficient between the sets of common subgraphs in Graph A and Graph B. The Jaccard distance is defined as 1 minus the Jaccard similarity, providing a measure of dissimilarity.

Edit Distance: Compute the edit distance between the sets of common subgraphs. The edit distance represents the minimum number of operations (e.g., insertions, deletions, substitutions) required to transform one set of subgraphs into the other.

Graph Diffusion Distance: Use diffusion-based methods to compare the structural similarity of common subgraphs. These methods take into account both the presence and connections of common subgraphs in the graphs.
Normalization (Optional):

Depending on the specific distance measure used, you may want to normalize the distance value to ensure that it falls within a defined range (e.g., [0, 1]) or has a meaningful interpretation in the context of your application.
Interpretation:

Interpret the distance measure in the context of your problem. A lower distance value indicates greater structural similarity between the two graphs, while a higher value implies greater dissimilarity.
